```python
# F분포를 사용한 등분산 검정
# 두 집단 분산의 비율을 계산해서 F통계량을 구하며, F통계량은 F분포를 따름
```

```python
from scipy.stats import levene

group1 = [12, 14, 15, 13, 12]
group2 = [22, 25, 27, 23, 21]

stat, p = levene(group1, group2)

print("Levene’s Test: stat =", stat, ", p =", p)
```

```python
# 정규 분포를 따른다는 가정이 필요
# 검정통계량은 자유도가 모집단 수 - 1 인 카이제곱분포를 따름
from scipy.stats import bartlett 

group1 = [12, 14, 15, 13, 12]
group2 = [22, 25, 27, 23, 21]

stat, p = bartlett(group1, group2)

print("Bartlett’s Test: stat =", stat, ", p =", p)
```

```python
from scipy.stats import fligner

group1 = [12, 14, 15, 13, 12]
group2 = [22, 25, 27, 23, 21]

stat, p = fligner(group1, group2)

print("Fligner-Killeen Test: stat =", stat, ", p =", p)
```

```python
import statsmodels.api as sm
import statsmodels.stats.api as sms
import numpy as np

# 예시 데이터
x = np.random.normal(0, 1, 100)
y = 2 * x + np.random.normal(0, 1, 100)

X = sm.add_constant(x)
model = sm.OLS(y, X).fit()

# Breusch-Pagan test
test_stat, p_value, _, _ = sms.het_breuschpagan(model.resid, model.model.exog)
print("Breusch-Pagan Test: stat =", test_stat, ", p =", p_value)
```

```python
from statsmodels.stats.diagnostic import het_white

white_test = het_white(model.resid, model.model.exog)

print("White’s Test: stat =", white_test[0], ", p =", white_test[1])
```
