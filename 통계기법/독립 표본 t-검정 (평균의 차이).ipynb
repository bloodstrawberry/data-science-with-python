# 신뢰구간



# 자유도



```python
import pandas as pd
import numpy as np

np.random.seed(1234)

sample0 = np.random.normal(loc=100, scale=10, size=50)
sample1 = np.random.normal(loc=95, scale=15, size=50)

df = pd.DataFrame({
    'sample0': sample0,
    'sample1': pd.Series(sample1)
})

df.head()
```

```python
sam0 = df['sample0']
sam1 = df['sample1']

num0 = len(sam0)
num1 = len(sam1)

mean0 = sam0.mean()
mean1 = sam1.mean()

num0, num1, mean0, mean1
```

```python
std0 = sam0.std()
std1 = sam1.std()

std0, std1
```

```python
# 신뢰구간 95% = 1.96

low = (mean0 - mean1) - 1.96 * np.sqrt((std0 ** 2 / num0) + (std1 ** 2 / num1))
up  = (mean0 - mean1) + 1.96 * np.sqrt((std0 ** 2 / num0) + (std1 ** 2 / num1))

(up - low)
```
