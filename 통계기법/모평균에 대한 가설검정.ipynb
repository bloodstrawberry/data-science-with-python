```python
from scipy.stats import norm

# Z  
z = 1.8

#  : p-value = 2 * P(Z > 1.8)
p_value = 2 * norm.sf(abs(z))  #  2 * (1 - norm.cdf(abs(z)))

print(f"Z = {z:.2f}")
print(f"p-value = {p_value:.4f}")
```

```python
import numpy as np
from scipy.stats import ttest_1samp

#     
scores = [68, 75, 70, 72, 69, 74, 71, 70, 73, 68]

#   
population_mean = 70

#   t- 
t_stat, p_value = ttest_1samp(scores, population_mean)

#  
print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_value:.3f}")

#  0.05  
alpha = 0.05
if p_value < alpha:
    print(" :     .")
else:
    print(" :      .")
```

```python
import numpy as np
from scipy.stats import ttest_ind

groupA = [70, 68, 75, 80, 72, 78, 74] # 
groupB = [75, 70, 78, 82, 74, 80, 76, 70, 79, 75] # 

np.mean(groupA), np.mean(groupB), np.median(groupA), np.median(groupB)

print("     t- ")
t_stat, p_value = ttest_ind(groupA, groupB, equal_var=True)
print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_value:.4f}")

print()

#  0.05  
if p_value < alpha:
    print(" :     .")
else:
    print(" :      .")
```

```python
import numpy as np
from scipy.stats import ttest_ind

groupA = [70, 68, 75, 80, 72, 78, 74] # 
groupB = [75, 70, 78, 82, 74, 80, 76, 70, 79, 75] # 

np.mean(groupA), np.mean(groupB), np.median(groupA), np.median(groupB)

print("   Welchs t- ")
t_stat, p_value = ttest_ind(groupA, groupB, equal_var=False)
print(f"t-statistic: {t_stat:.3f}")
print(f"p-value: {p_value:.4f}")

print()

#  0.05  
if p_value < alpha:
    print(" :     .")
else:
    print(" :      .")
```
