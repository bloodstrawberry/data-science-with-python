```python
import pandas as pd
import numpy as np

np.random.seed(1234)

num_rows = 1000
product_choices = ['PHONE', 'AIR', 'MONITOR', 'USB']

data = {
    'ID1': np.random.randint(1, 5, size=num_rows),  # 고객 ID: 1~20 사이
    'ID2': np.random.randint(1, 5, size=num_rows),  # 고객 ID: 1~20 사이    
    'PRODUCT': np.random.choice(product_choices, size=num_rows),
    'PRICE1': np.random.randint(100, 2001, size=num_rows),  # 가격: 100 ~ 2000원 사이
    'PRICE2': np.random.randint(100, 2001, size=num_rows)  # 가격: 100 ~ 2000원 사이    
}

df = pd.DataFrame(data)

df 
```

```python
# ID를 행의 Key로 
# PRODUCT의 고유한 값들을 컬럼 명으로
# PRICE를 각 컬럼의 값으로 가지도록
# 데이터 구조 변경
pivot_df = pd.pivot_table(df, index='ID1', values='PRICE1', columns='PRODUCT')

# pivot_df = pivot_df.reset_index().rename_axis(None, axis=1)

# 출력
# pivot_df = pivot_df.loc[:, ['AIR', 'PHONE', 'USB', 'MONITOR']]
pivot_df
```
