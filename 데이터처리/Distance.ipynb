```python
# Manhattan Distance
# Euclidean Distance
# Minkowski Distance
# Mahalanobis Distance
```

```python
from scipy.spatial import distance
import numpy as np

p = np.array([2, 4, 6])
q = np.array([5, 1, 3])

p, q
```

```python
print("Euclidean:", distance.euclidean(p, q))
print("Manhattan:", distance.cityblock(p, q))
```

```python
print("Minkowski (p=3):", distance.minkowski(p, q, p=3))
```

```python
from sklearn.metrics import pairwise_distances

X = np.array([[2, 4, 6], [5, 1, 3]])

print("Euclidean matrix:\n", pairwise_distances(X, metric='euclidean'))
print("Manhattan matrix:\n", pairwise_distances(X, metric='manhattan'))
print("Minkowski (p=3) matrix:\n", pairwise_distances(X, metric='minkowski', p=3))
```
