```python
# Random Under Sampling
# 다수 클래스에서 데이터를 무작위로 제거.
# 장점: 간단하고 빠름.
# 단점: 중요한 정보가 손실될 수 있음 (랜덤하게 제거하기 때문에).
```

```python
# Tomek Links
# 두 클래스의 경계에 있는 샘플 쌍 중, 서로 가장 가까운 두 샘플을 찾아 제거.
# Tomek link의 정의: 서로 가장 가까운 이웃인데, 서로 다른 클래스에 속해 있을 때.
# 보통 다수 클래스의 샘플만 제거하여 클래스 분리를 명확히 함.
```

```python
# Random Over Sampling (Resampling)
# 소수 클래스의 데이터를 복제하여 증가.
# Adasyn, smote는 랜덤 오버샘플링의 고도화 방법은 아님 (이 둘은 소수 클래스의 데이터를 합성하는 기법, 접근 방식이 다름)

# 장점: 간단함.
# 단점: 과적합(overfitting) 가능성이 있음 (복제이므로).
```

```python
# SMOTE (Synthetic Minority Over-sampling Technique)
# 소수 클래스 샘플에 대해 주변 소수 클래스 샘플과 선형보간으로 데이터를 생성
# 소수 데이터들 사이를 보간하여 작동하기 때문에 새로운 사례의 데이터 예측엔 취약
# 인공적으로 새로운 데이터를 만들어 과적합을 줄임.
# KNN을 이용하여 Minority에서 새로운 샘플을 생성

# 소수 클래스의 수가 충분하면 새로운 샘플을 생성하지 않을 수도 있다. (e.g., sampling_strategy로 조절 가능)

```

```python
# ADASYN (Adaptive Synthetic Sampling)
# SMOTE의 확장형.
# 분류하기 어려운 소수 클래스 근처에 더 많은 synthetic 샘플을 생성함.
# 불균형 비율에 적응적으로 대응함.
```

```python
# MCMC (Markov Chain Monte Carlo)

# MCMC는 샘플링 기법이며, 위의 언더/오버샘플링과는 다른 목적(통계적 추정, 베이지안 추론)에 사용됩니다.

# 개요:
# 복잡한 확률분포에서 표본을 추출하고자 할 때 사용하는 방법.
# 직접적으로 분포를 알 수 없지만, 표본을 뽑아서 추론 또는 예측에 사용.

# 핵심 아이디어:
# Markov Chain: 현재 상태만으로 다음 상태를 결정하는 확률 과정.
# Monte Carlo: 확률 기반 시뮬레이션 기법.
# MCMC는 Markov Chain을 구성하여 많은 표본을 생성한 뒤, 이를 사용해 복잡한 분포를 근사함.

# 대표 알고리즘:
# Metropolis-Hastings
# Gibbs Sampling

# 활용 예:
# 베이지안 추론에서 사후분포 샘플링
# 머신러닝의 확률적 모델 추정
# 강화학습, 자연과학 모사 등 다양한 영역
```
