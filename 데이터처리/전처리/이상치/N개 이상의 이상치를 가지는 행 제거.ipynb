```python
import pandas as pd
import numpy as np

# 랜덤 데이터 생성
df = pd.DataFrame(
    np.random.randint(1, 11, size=(100, 5)),
    columns=['A', 'B', 'C', 'D', 'E']
)

df.head()
```

```python
# A, B, C, D, E 컬럼 중 3개 이상의 값이 7이상이면 이상치일 때, 이상치 제거
value_check = (df >= 7) * 1
value_check
```

```python
outlier_count = value_check.sum(axis=1)
outlier_count
```

```python
outlier_idx = outlier_count[outlier_count >= 3].index.tolist()
outlier_idx
```

```python
# 이상치인 index를 제외
df_new = df[df.index.isin(outlier_idx) == False]
df_new
```

```python
# 이상치인 index를 drop
df_new = df.drop(outlier_idx) # axis=0
df_new
```
