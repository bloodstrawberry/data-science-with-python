```python
# 데이터 추가
```

```python
import pandas as pd
import numpy as np

# 랜덤 데이터 생성
df = pd.DataFrame(
    np.random.randint(1, 11, size=(5, 5)),
    columns=['A', 'B', 'C', 'D', 'E']
)

df.head()
```

```python
# 데이터 추가 - loc
df.loc[5] = [1, 2, 3, 4, 5]
df
```

```python
# 데이터 추가 - dictionary
df.loc[6] = {'A' : 1, 'B' : 2, 'C' : -1, 'D' : -5, 'E' : -6}
df
```

```python
# 값 변경
df.loc[0, 'A'] = 'HELLO'
df.iloc[0, 1] = 'WORLD'

df
```

```python
df.replace(1, 100)
```

```python
df['B'].replace({3 : 300, 2 : 200})
```
