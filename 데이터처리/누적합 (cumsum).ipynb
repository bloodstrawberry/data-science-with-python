```python
import pandas as pd

# 예시 데이터프레임
df = pd.DataFrame({
    'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'sales': [100, 150, 200, 130, 170]
})

df
```

```python
# 구현

df['cum'] = 0
for i in range(df.shape[0]):
    df.loc[i, 'cum'] = df.loc[:i, 'sales'].sum()

df
```

```python
# cumsum

df['cumulative_sales'] = df['sales'].cumsum()

df
```
