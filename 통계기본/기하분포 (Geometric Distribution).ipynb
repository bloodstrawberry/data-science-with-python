```python
# X ~ Geometric(p)

# 성공 확률이 p인 베르누이 시행을 반복할 때,
# 처음 성공이 나올 때까지 시행한 횟수를 확률 변수 X라고 하면,
# 이 X는 기하분포를 따른다.

# 첫 성공이 나올 때까지 시행한 횟수 : X = 1, 2, 3, ...
# 첫 성공이 나올 때까지의 실패 횟수 : X = 0, 1, 2, ...
```

### 첫 성공이 x번째 시행에서 나올 확률



```python
# 성공 확률이 0.2라면 E(X) = 1 / p = 1 / 0.2 = 5 -> 5번에 1번은 성공한다는 의미가 된다.
```

```python
# 성공 확률이 0.2일 때, 첫번째에 실패하고 두번째에 성공할 확률은?
p = 0.2
q = (1 - p)

p * q # P(X = 2)
```

```python
# 5번보다 적은 횟수 안에 성공할 확률은?
1 - q ** 5 # P(X <= 5)
```

```python
# 초콜릿 안에 20% 확률로 골든 티켓이 있을 때, 골든 티켓을 처음으로 발견하기 전까지 4개의 초콜릿보다 적은 초콜릿을 열어볼 확률은?
p = 0.2
mu = 1 / p
var = (1 - p) / (p ** 2)

mu, var
```

```python
# P(X <= 3)
1 - (1 - p) ** 3
```
