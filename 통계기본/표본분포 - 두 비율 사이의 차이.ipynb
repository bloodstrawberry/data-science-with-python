```python
import numpy as np
from scipy import stats

# 입력: 성공 횟수 x, 전체 시도 n
x1, n1 = 45, 100   # 그룹 A
x2, n2 = 30, 90    # 그룹 B

# 표본 비율 계산
p1 = x1 / n1
p2 = x2 / n2
diff = p1 - p2

# 표준오차 계산
se = np.sqrt((p1 * (1 - p1)) / n1 + (p2 * (1 - p2)) / n2)

# z 통계량
z_stat = diff / se

# 양측 검정 p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# 신뢰구간 (95%)
z_critical = stats.norm.ppf(0.975)  # 약 1.96
ci_low = diff - z_critical * se
ci_high = diff + z_critical * se

# 출력
print(f"그룹 A 성공률: {p1:.3f}, 그룹 B 성공률: {p2:.3f}")
print(f"비율 차이: {diff:.3f}")
print(f"표준오차(SE): {se:.4f}")
print(f"z 통계량: {z_stat:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"95% 신뢰구간: ({ci_low:.4f}, {ci_high:.4f})")

# 해석
if ci_low > 0 or ci_high < 0:
    print("✅ 신뢰구간이 0을 포함하지 않음 → 통계적으로 유의한 차이 있음.")
else:
    print("❌ 신뢰구간에 0 포함 → 통계적으로 유의한 차이 없음.")
```
