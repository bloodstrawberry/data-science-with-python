```python
import numpy as np
from scipy import stats

# :   x,   n
x1, n1 = 45, 100   #  A
x2, n2 = 30, 90    #  B

#   
p1 = x1 / n1
p2 = x2 / n2
diff = p1 - p2

#  
se = np.sqrt((p1 * (1 - p1)) / n1 + (p2 * (1 - p2)) / n2)

# z 
z_stat = diff / se

#   p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

#  (95%)
z_critical = stats.norm.ppf(0.975)  #  1.96
ci_low = diff - z_critical * se
ci_high = diff + z_critical * se

# 
print(f" A : {p1:.3f},  B : {p2:.3f}")
print(f" : {diff:.3f}")
print(f"(SE): {se:.4f}")
print(f"z : {z_stat:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"95% : ({ci_low:.4f}, {ci_high:.4f})")

# 
if ci_low > 0 or ci_high < 0:
    print("  0       .")
else:
    print("  0      .")
```
