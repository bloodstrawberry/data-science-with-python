```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 평균과 표준편차
mu, sigma = 0, 1

# x 값 범위
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, mu, sigma)

# 구간 설정
x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-2, 2, 100)
x3 = np.linspace(-3, 3, 100)

# 정규분포 PDF 그리기
plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='black')

# 경험적 법칙 구간 색칠
plt.fill_between(x1, norm.pdf(x1, mu, sigma), alpha=0.4, label='±1σ ≈ 68%', color='skyblue')
plt.fill_between(x2, norm.pdf(x2, mu, sigma), alpha=0.4, label='±2σ ≈ 95%', color='orange')
plt.fill_between(x3, norm.pdf(x3, mu, sigma), alpha=0.4, label='±3σ ≈ 99.7%', color='green')

# 시각적 설정
plt.title('(68-95-99.7 Rule)', fontsize=14)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 체비셰프의 법칙과 경험적 법칙 비교 시각화

plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='black')

# 체비셰프 하한선 대응 색상으로 음영 처리 (보수적 기준)
plt.axvspan(-2, 2, alpha=0.2, color='purple', label='Chebyshev: ≥75% within ±2σ')
plt.axvspan(-3, 3, alpha=0.2, color='red', label='Chebyshev: ≥88.9% within ±3σ')

# 경험적 법칙 색상 다시 추가
plt.fill_between(x1, norm.pdf(x1, mu, sigma), alpha=0.4, color='skyblue', label='Empirical: ≈68% within ±1σ')
plt.fill_between(x2, norm.pdf(x2, mu, sigma), alpha=0.4, color='orange', label='Empirical: ≈95% within ±2σ')
plt.fill_between(x3, norm.pdf(x3, mu, sigma), alpha=0.4, color='green', label='Empirical: ≈99.7% within ±3σ')

# 시각적 설정
plt.title('Normal Distribution vs Chebyshev Theorem', fontsize=14)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```
