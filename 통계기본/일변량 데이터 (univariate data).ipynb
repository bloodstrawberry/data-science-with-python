```python
# 일변량 데이터 : 단일 변수만을 포함하는 데이터 (어떤 반의 학생들의 키)

# 단변량 수치형 데이터(Univariate Numerical Data) : 하나의 수치형 변수만으로 구성된 데이터 (시험 점수, 수면 시간)
# 대표 차트 : 줄기와 잎 그림 (Stem-and-Leaf Plot), 바이올린 플롯, 히스토그램, 박스 플롯
```

```python
# stemgraphic 모듈 없이 줄기와 잎 그림을 텍스트로 구현하고, 바이올린 플롯은 시각화

import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict

# 단변량 수치형 데이터 예시: 시험 점수
scores = [72, 75, 78, 82, 85, 85, 86, 87, 88, 89, 90, 91, 92, 93, 95]

# 줄기와 잎 그림 (텍스트 기반)
stem_leaf = defaultdict(list)
for score in scores:
    stem = score // 10
    leaf = score % 10
    stem_leaf[stem].append(leaf)

# 줄기와 잎 텍스트 출력 준비
stem_leaf_output = "stem | leaf\n"
for stem in sorted(stem_leaf.keys()):
    leaves = " ".join(str(leaf) for leaf in sorted(stem_leaf[stem]))
    stem_leaf_output += f" {stem}  | {leaves}\n"

# 바이올린 플롯 시각화
sns.set(style="whitegrid")
plt.figure(figsize=(6, 5))
sns.violinplot(y=scores, color='skyblue')
plt.title("violin plot")
plt.ylabel("score")
plt.tight_layout()
plt.show()

stem_leaf_output
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 일변량 데이터: 키
heights = [160, 165, 167, 162, 170, 168, 172, 166, 164, 169]

# 그래프 스타일 설정
sns.set(style="whitegrid")

# 1행 2열 subplot 생성
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# 히스토그램
sns.histplot(heights, bins=5, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('histogram')
axes[0].set_xlabel('height (cm)')
axes[0].set_ylabel('mode')

# 박스플롯
sns.boxplot(y=heights, ax=axes[1], color='lightgreen')
axes[1].set_title('box plot')
axes[1].set_ylabel('height (cm)')

plt.tight_layout()
plt.show()
```
