```python
# 최대 우도 추정법
# MLE, Maximum Likelihood Estimator)
# 모수(parameter)를 추정하는 방법 중 하나
# 현재 관측된 데이터가 가장 일어날 법하게 만들어주는 θ
# 주어진 데이터에서 관측될 가능성이 가장 높은 모수(θ)를 추정하는 방법
```

```python
# 확률 : 분포가 정해지고, 분포의 모수가 정해졌을 때 관측치가 나올 가능성
# 우도 : 분포가 정해지고, 관측치가 주어졌을 때 모수가 나올 가능성
```

```python
import numpy as np
import matplotlib.pyplot as plt

# theta 값 범위
theta = np.linspace(0, 1, 500)

# 우도 함수: L(theta) = 120 * theta^7 * (1 - theta)^3
likelihood = 120 * theta**7 * (1 - theta)**3

# 최대 우도 지점
theta_mle = 0.7
likelihood_mle = 120 * theta_mle**7 * (1 - theta_mle)**3

# 그래프 그리기
plt.figure(figsize=(8, 5))
plt.plot(theta, likelihood, label=r'$L(\theta) = 120 \cdot \theta^7 (1 - \theta)^3$', color='blue')
plt.axvline(theta_mle, color='red', linestyle='--', label=r'MLE: $\hat{\theta} = 0.7$')
plt.scatter([theta_mle], [likelihood_mle], color='red')
plt.title('Likelihood Function for Binomial(10, θ), X = 7')
plt.xlabel(r'$\theta$')
plt.ylabel('Likelihood')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
