```python
# X ~ Binomial(n, p)

# 성공 확률이 p인 동일한 베르누이 시행을 총 n번 독립적으로 반복했을 때,
# 성공한 횟수 X를 확률 변수로 하는 분포
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# 공통 설정
n = 20  # 시행 횟수
x = np.arange(0, n + 1)  # 가능한 성공 횟수

# 각 확률에 대한 이항분포 PMF
ps = [0.3, 0.5, 0.7]
colors = ['blue', 'green', 'red']
labels = ['p = 0.3 (p < 0.5)', 'p = 0.5', 'p = 0.7 (p > 0.5)']

# 그림 그리기
plt.figure(figsize=(10, 6))
for p, color, label in zip(ps, colors, labels):
    pmf = binom.pmf(x, n, p)
    plt.plot(x, pmf, marker='o', linestyle='-', color=color, label=label)

plt.title('Binomial Distribution (n = 20) for Different p')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability P(X = k)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
