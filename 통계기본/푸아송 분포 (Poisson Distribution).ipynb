### 독립적인 푸아송 분포 X ~ Poisson(λx), Y ~ Poisson(λy)에 대해
#### X, Y과 독립확률변수라면 P(X + Y) = P(X) + P(Y), E(X + Y) = E(X) + E(Y)




```python
# 이항분포 X ~ B(n, p)에서 E(X) = np, V(X) = npq가 λ와 비슷한 경우,
# 즉, n이 크고 q가 1에 가까우면 X ~ B(n, p)는 X ~ Po(np)에 근사할 수 있다.
```

```python
# 성공확률이 0.05일 때, 50번 시도해서 5번 성공할 확률은?
from scipy.stats import binom # 이항분포
from scipy.stats import poisson # 푸아송 분포

n = 50
p = 0.05
np = 2.5

ld = np

binom.pmf(5, n, p), poisson.pmf(5, ld)
```

```python
# 15분마다 택시가 지나갈 때, 15분 동안 한 대도 지나가지 않을 확률은?
from scipy.stats import poisson

ld = 1
poisson.pmf(0, ld)
```

```python
# 어느 기관에서 시간당 10건의 전화를 받는다. 5분 내에 1건 이상 전화 받을 확률은?
from scipy.stats import poisson

# 시간당 10건 -> 60분당 10건 
# 평균 발생 건수 (lambda): 5분 동안 평균 0.8333건
lambda_5min = 10 * (5 / 60)

# P(X >= 1) = 1 - P(X = 0)
prob_at_least_one = 1 - poisson.pmf(0, lambda_5min)

print(f"5분 내에 1건 이상 전화 받을 확률: {prob_at_least_one:.4f}")
```

```python
# 이항분포 B(n=100, p=0.02)를 포아송 분포로 근사하여 X = 0일 확률?
from scipy.stats import poisson

n = 100
p = 0.02
ld = n * p

poisson.pmf(0, ld)
```

```python
# 어떤 버스는 시간 당 평균 5대의 호출 요청을 받을 때, 시간 당 3대 이상의 호출을 받을 확률은?
from scipy.stats import poisson

# 평균 호출 수 (lambda)
lam = 5

# P(X >= 3) = 1 - P(X <= 2)
prob = 1 - poisson.cdf(2, lam)

print(f"시간당 3대 이상의 호출을 받을 확률: {prob:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# 설정
lambdas = [1, 4.5, 10]  # 평균 λ 값들
x = np.arange(0, 21)  # x축 범위

# 그래프 그리기
plt.figure(figsize=(10, 6))
for lam in lambdas:
    pmf = poisson.pmf(x, mu=lam)
    plt.plot(x, pmf, marker='o', linestyle='-', label=f'λ = {lam}')

plt.title('Poisson Distribution for Different λ Values')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability P(X = k)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
