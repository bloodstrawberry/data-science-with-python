```python
MLE(Maximum Likelihood Estimation, 최대우도추정법)의 주요 특징은 다음과 같습니다:

✅ 1. 데이터에 가장 잘 맞는 파라미터 추정
관측된 데이터를 가장 잘 설명하는(우도 likelihood가 최대가 되는) 모수(parameter) 를 찾는 방법입니다.

즉, 데이터가 주어졌을 때 그 데이터가 나올 확률을 최대화하는 파라미터를 찾습니다.

✅ 2. 점추정(point estimation) 방식
확률적이지 않은 하나의 고정된 값을 추정합니다.

베이지안 방법처럼 분포(사후확률분포)를 추정하지 않습니다.

✅ 3. 대수우도함수(log-likelihood)를 자주 사용
계산의 편의성과 수치 안정성을 위해 우도함수에 로그를 취한 로그 우도함수(log-likelihood) 를 최적화합니다.

✅ 4. 일관성(Consistency)
샘플 수가 충분히 많아지면, MLE 추정값은 실제 모수값에 수렴합니다 (확률적으로 수렴).

✅ 5. 점근적 정규성(Asymptotic normality)
표본 수가 무한히 커질 때, MLE는 정규분포를 따릅니다.

즉, 추정치의 분포는 정규분포로 근사 가능하며, 신뢰구간 계산에 유리합니다.

✅ 6. 효율성(Efficiency)
샘플 수가 커질수록, MLE는 최소 분산 추정량(Cramér-Rao Bound를 달성) 이라는 점에서 효율적입니다.

✅ 7. 충분통계량(sufficient statistic)을 잘 활용
MLE는 충분통계량이 존재할 경우, 그것만으로도 동일한 추정을 할 수 있어 차원 축소에 유리합니다.

✅ 8. 사전확률 사용 X (비베이지안 접근)
MLE는 사전확률(prior) 을 고려하지 않고, 오직 데이터로부터 추정합니다.

베이즈 추정과 대비되는 특징입니다.

✅ 9. 최적화 문제로 귀결
MLE는 결국 우도 또는 로그우도함수를 최대화하는 문제이므로, 경사하강법, 뉴턴 방법, 수치 최적화 알고리즘 등을 사용합니다.
```

```python
🔷 1. MLE 개념: "관측된 데이터가 가장 그럴듯하게 나올 확률"
MLE는 **"주어진 데이터가 실제로 관측될 확률이 가장 높게 되도록 하는 파라미터를 찾는 방법"**입니다.

즉, 어떤 분포를 따르는 데이터에서 모수 θ를 추정할 때:

"데이터가 실제로 이렇게 나왔다는 사실을 가장 잘 설명하는 θ는 뭐지?"
→ 그 θ가 바로 MLE로 추정된 값입니다.
```
