```python
# 신경망 예시
# 4 layer 신경망
# 입력층 (Layer 1)
# 은닉층 2개 (Layer 2, 3)
# 출력층 (Layer 4)
# 보통 layer라고 하면 입력층도 포함해서 4개층인 구조

# 4-hidden layer 신경망
# 입력층 (Layer 1)
# 은닉층 4개 (Layer 2, 3, 4, 5)
# 출력층 (Layer 6)
# 총 6개 층.

# Fully connected feed forward 신경망
# 입력층부터 출력층까지 모든 층이 앞 층과 완전 연결된 (Fully connected) 순방향 신경망

# 5-layer 신경망
# 입력층 (1)
# 은닉층 3개 (2,3,4)
# 출력층 (5)
```

```python
import matplotlib.pyplot as plt

def draw_neural_net(ax, left, right, bottom, top, layer_sizes):
    '''
    간단히 층별 노드와 연결을 그리는 함수
    layer_sizes: 각 층별 노드 수 리스트
    '''
    v_spacing = (top - bottom) / float(max(layer_sizes))
    h_spacing = (right - left) / float(len(layer_sizes) - 1)
    
    # 노드 위치 계산
    for i, layer_size in enumerate(layer_sizes):
        layer_top = v_spacing * (layer_size - 1) / 2. + (top + bottom)/2.
        for j in range(layer_size):
            circle = plt.Circle((left + i * h_spacing, layer_top - j * v_spacing), 0.03, 
                                color='w', ec='k', zorder=4)
            ax.add_artist(circle)
    
    # 연결선 그리기
    for i, (layer_size_a, layer_size_b) in enumerate(zip(layer_sizes[:-1], layer_sizes[1:])):
        layer_top_a = v_spacing * (layer_size_a - 1) / 2. + (top + bottom)/2.
        layer_top_b = v_spacing * (layer_size_b - 1) / 2. + (top + bottom)/2.
        for j in range(layer_size_a):
            for k in range(layer_size_b):
                ax.plot([left + i * h_spacing, left + (i + 1) * h_spacing],
                        [layer_top_a - j * v_spacing, layer_top_b - k * v_spacing], 'k-', lw=0.5)

    ax.axis('off')

fig, axs = plt.subplots(2, 2, figsize=(12,10))

# 1) 4 layer 신경망 (입력 3, 은닉 4, 은닉 4, 출력 2)
axs[0,0].set_title("4 Layer Neural Network")
draw_neural_net(axs[0,0], .1, .9, .1, .9, [3,4,4,2])

# 2) 4-hidden layer 신경망 (입력 3, 은닉 5,5,5,5, 출력 2)
axs[0,1].set_title("4 Hidden Layer Neural Network")
draw_neural_net(axs[0,1], .1, .9, .1, .9, [3,5,5,5,5,2])

# 3) Fully connected feed forward 신경망 (입력 3, 은닉 4, 출력 2)
axs[1,0].set_title("Fully Connected Feed Forward NN")
draw_neural_net(axs[1,0], .1, .9, .1, .9, [3,4,2])

# 4) 5-layer 신경망 (입력 3, 은닉 4,4,4, 출력 2)
axs[1,1].set_title("5 Layer Neural Network")
draw_neural_net(axs[1,1], .1, .9, .1, .9, [3,4,4,4,2])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def binary_step(x):
    return np.where(x >= 0, 1, 0)

x = np.linspace(-5, 5, 500)
y = binary_step(x)

plt.plot(x, y, drawstyle='steps-post')
plt.title('Binary Step Function')
plt.xlabel('x')
plt.ylabel('output')
plt.grid(True)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 입력 값 범위 설정
x = np.linspace(-10, 10, 400)

# 활성화 함수 정의
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# 각 함수 적용
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)

# 4개의 서브플롯으로 분할하여 개별 활성화 함수 시각화
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Sigmoid
axs[0, 0].plot(x, y_sigmoid, color='blue')
axs[0, 0].set_title("Sigmoid")
axs[0, 0].grid(True)
axs[0, 0].axhline(0, color='black', linewidth=0.5)
axs[0, 0].axvline(0, color='black', linewidth=0.5)

# Tanh
axs[0, 1].plot(x, y_tanh, color='green')
axs[0, 1].set_title("Tanh")
axs[0, 1].grid(True)
axs[0, 1].axhline(0, color='black', linewidth=0.5)
axs[0, 1].axvline(0, color='black', linewidth=0.5)

# ReLU
axs[1, 0].plot(x, y_relu, color='red')
axs[1, 0].set_title("ReLU")
axs[1, 0].grid(True)
axs[1, 0].axhline(0, color='black', linewidth=0.5)
axs[1, 0].axvline(0, color='black', linewidth=0.5)

# Leaky ReLU
axs[1, 1].plot(x, y_leaky_relu, color='purple')
axs[1, 1].set_title("Leaky ReLU")
axs[1, 1].grid(True)
axs[1, 1].axhline(0, color='black', linewidth=0.5)
axs[1, 1].axvline(0, color='black', linewidth=0.5)

# 전체 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
# 활성화 함수
# sigmoid : 로지스틱 함수와 동일
# tanh : 입력을 -1과 1 사이로 스케일링하며, zero-centered 특성을 가짐
# ReLU : 계산이 간단하고 빠름
# Leaky ReLU : ReLU의 대부분 출력이 0이 되는 뉴런이 죽는 문제(Dying ReLU)를 보완

# sigmoid와 tanh는 층이 깊어짐에 따라 동일하게 Vanishing Gradient 문제가 발생한다.
# sigmoid와 tanh는 적합한 가중치의 초기화 방법이 같다. (Xavier Normal Uniform)
# ReLU와 Leaky ReLU도 적합한 가중치의 초기화 방법이 같다. (He Normal, Uniform)
```
