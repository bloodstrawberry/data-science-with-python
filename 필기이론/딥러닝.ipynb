```python
# 3개의 서로 다른 클래스(Class)를 구별할 수 있는 신경망 모델
# 다층 퍼셉트론 (MLP) : 비선형 활성화 함수와 다층 구조를 이용해 복잡한 함수 근사 및 다중 클래스 분류 가능.
# 순환 신경망 (RNN) : 시퀀스 데이터를 처리하기 위해 설계되었으며, 다중 클래스 분류도 가능. (텍스트 분류 감정 분석 = 긍정/중립/부정 등)
# 컨볼루션 신경망 (CNN) : 이미지, 시계열 등 다양한 입력에서 패턴을 추출하고, 다중 클래스 분류에 매우 효과적.
```

### 순환 신경망 모델 구조 



```python
# one to one : 이미지 -> 고정된 크기의 이미지 분류
# one to many : 이미지 -> 이미지를 설명하는 문장 생성
# many to one : 문장의 단어 시퀀스 -> 긍정 or 부정
# many to many(1) : 주식 시장 변동 사항, 다음 주의 주식값 예측
# many to many(2) : 음성 신호에서 호출 여부를 주기마다 알려주는 시퀀스 모델 (Hey Google~ 이라고 하면 반응하여 구동하는 앱)
```
